```python
'''
1. Sort 함수를 작성하시오
array = [1,4,6,8,9,3,2,10]
size = 8
Sort(array,size)
'''

def Sort(array, size):
    cnt = 0

    while cnt < size:
        MIN = float('inf')
        idx = cnt

        for i in range(cnt, size):
            if MIN > array[i]:
                MIN = array[i]
                idx = i
        array[cnt], array[idx] = array[idx], array[cnt]
        cnt += 1

    return array


```

```python
array = [1,4,6,8,9,3,2,10]
size = 8
Sort(array, size)
```

```python
'''
2. 다음과 같은 모양을 출력할 수 있도록(size = 5)
    *
   * *
  *   *
 *     *
*********
'''
# GPT
size = int(input())

for i in range(size):
    if i == size - 1:
        print('*' * (2 * size - 1))
    else:
        left_space = ' ' * (size - i - 1)

        if i == 0:
            print(left_space + '*')
        else:
            inner_space = ' ' * (2 * i - 1)
            print(left_space + '*' + inner_space + '*')
```
