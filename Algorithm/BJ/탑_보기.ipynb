## 코드

```python
import sys
input = input

N = int(input())
H = [0] + list(map(int, input().split()))

cnt = [0] * (N + 1)
near = [0] * (N + 1)

# 1️⃣ 왼쪽 → 오른쪽
stack = []
for i in range(1, N + 1):
    while stack and stack[-1][1] <= H[i]:
        stack.pop()

    cnt[i] += len(stack)

    if stack:
        near[i] = stack[-1][0]

    stack.append((i, H[i]))

# 2️⃣ 오른쪽 → 왼쪽
stack = []
for i in range(N, 0, -1):
    while stack and stack[-1][1] <= H[i]:
        stack.pop()

    cnt[i] += len(stack)

    if stack:
        # 후보 비교
        if near[i] == 0:
            near[i] = stack[-1][0]
        else:
            left = near[i]
            right = stack[-1][0]

            if abs(i - right) < abs(i - left):
                near[i] = right
            elif abs(i - right) == abs(i - left):
                near[i] = min(left, right)

    stack.append((i, H[i]))

# 출력
for i in range(1, N + 1):
    if cnt[i] == 0:
        print(0)
    else:
        print(cnt[i], near[i])
```
