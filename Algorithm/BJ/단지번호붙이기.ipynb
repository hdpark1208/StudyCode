## 구상

- dfs 방식
- 단지를 어떻게 구분지을까
    - 전체를 다 돌면서 방문처리 안된 "1"을 찾아야지



## 코드

```python
import sys
import io

sys.stdin = io.StringIO("""7
0110100
0110101
1110101
0000111
0100000
0111110
0111000
""")

N = int(sys.stdin.readline())
graph = [list(map(int, sys.stdin.readline().rstrip())) for _ in range(N)]

# N = int(input())
# graph = []
# for _ in range(N):
#     graph.append(list(map(int, input())))

visited = [[False]*N for _ in range(N)]
dirs = [(-1,0),(1,0),(0,-1),(0,1)]

def dfs(x, y):
    visited[x][y] = True
    cnt = 1
    for dx, dy in dirs:
        nx, ny = x + dx, y + dy
        if 0 <= nx < N and 0 <= ny < N:
            if graph[nx][ny] == 1 and not visited[nx][ny]:
                cnt += dfs(nx, ny)
    return cnt

complexes = []
for i in range(N):
    for j in range(N):
        if graph[i][j] == 1 and not visited[i][j]:
            complexes.append(dfs(i, j))

print(len(complexes))
for x in sorted(complexes):
    print(x)
```

```python
# Fail
N = int(input())
graph = []
for _ in range(N):
    graph.append(list(map(int, input())))

visited = [[False] * N for _ in range(N)]
result_graph = [[0] * N for _ in range(N)]

def dfs(x, y):
    global village

    if graph[x][y] == 0:
        return -1

    if graph[x][y] == 1 and visited[x][y] == False:
        visited[x][y] = True
        village += 1
        result_graph[x][y] = village

    dirs = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    for dx, dy in dirs:
        nx = x + dx
        ny = y + dy
        if 0 <= nx < N and 0 <= ny < N and visited[nx][ny] == False:
            visited[nx][ny] = True
            dfs(nx, ny)
            if graph[nx][ny] == 1:
                result_graph[x][y] = village

    return -1

village = 0
for i in range(N):
    for j in range(N):
        dfs(i, j)

print(village)
print(result_graph)
```
