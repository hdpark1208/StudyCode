## 코드

```python
from collections import deque

N, M, K, X = map(int, input().split())

graph = [[] for _ in range(N + 1)]

for _ in range(M):
    a, b = map(int, input().split())
    graph[a].append(b)

dist = [-1] * (N + 1)
dist[X] = 0

q = deque([X])

while q:
    now = q.popleft()

    for nxt in graph[now]:
        if dist[nxt] == -1:
            dist[nxt] = dist[now] + 1
            q.append(nxt)

found = False

for city in range(1, N + 1):
    if dist[city] == K:
        print(city)
        found = True

if not found:
    print(-1)
```

```python
# 다익스트라

import heapq

N, M, K, X = map(int, input().split())

graph = [[] for _ in range(N + 1)]

for _ in range(M):
    a, b = map(int, input().split())
    graph[a].append((b, 1))  # 비용 1

INF = 10**15
dist = [INF] * (N + 1)
dist[X] = 0

pq = []
heapq.heappush(pq, (0, X))  # (거리, 정점)

while pq:
    cur_dist, now = heapq.heappop(pq)

    # 이미 더 짧은 거리로 처리된 적 있으면 무시
    if cur_dist > dist[now]:
        continue

    for nxt, cost in graph[now]:
        new_dist = cur_dist + cost

        if new_dist < dist[nxt]:
            dist[nxt] = new_dist
            heapq.heappush(pq, (new_dist, nxt))

found = False

for city in range(1, N + 1):
    if dist[city] == K:
        print(city)
        found = True

if not found:
    print(-1)
```
