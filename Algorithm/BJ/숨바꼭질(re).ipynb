## 구상

- BFS
- BFS의 정답 테이블(방문처리 겸)이 최소 시간을 보장한다

## 코드

```python
from collections import deque

N, K = map(int, input().split())
MAX = 100000

dist = [-1] * (MAX + 1)
dist[N] = 0

q = deque([N])

while q:
    x = q.popleft()
    if x == K:
        break

    for nx in (x - 1, x + 1, x * 2):
        if 0 <= nx <= MAX and dist[nx] == -1:
            dist[nx] = dist[x] + 1
            q.append(nx)

print(dist[K])
```
