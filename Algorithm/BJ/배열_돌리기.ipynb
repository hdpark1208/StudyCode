## 구상

- 주 대각선, 중간 열, 부대각선, 중간 행 리스트를 따로 빼 둔 후, 각도에 맞춰서 배열에 끼워넣어보자



## 코드

```python
# 주 대각선: (1,1), (2,2), ..., (n,n)
# 중간 열: (1, n//2 + 1), (2, n//2 + 1), ..., (n, n//2 + 1)
# 부 대각선: (1, n), (2, n-1), (3, n-2), ..., (n, 1)
# 중간 행: (n//2 + 1, n), (n//2 + 1, n-1), ..., (n//2 + 1, 1)
```

```python
T = int(input())
for _ in range(T):
    n, d = map(int,input().split())
    graph = []

    diag_1, diag_2 = [], []
    mid_col, mid_row = [], []
    for _ in range(n):
        graph.append(list(map(int,input().split())))

    for i in range(n):
        diag_1.append(graph[i][i])
        mid_col.append(graph[i][n//2])
        diag_2.append(graph[i][n - i - 1])
        mid_row.append(graph[n//2][n - i - 1])

    k = (d // 45) % 8

    for _ in range(k):
        diag_1, mid_col, diag_2, mid_row = (
        mid_row[::-1],
        diag_1,
        mid_col,
        diag_2
    )

    for i in range(n):
        graph[i][i] = diag_1[i]
        graph[i][n // 2] = mid_col[i]
        graph[i][n - i - 1] = diag_2[i]
        graph[n // 2][n - i - 1] = mid_row[i]

    for row in graph:
        print(*row)

```
