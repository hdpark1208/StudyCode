## 구상

* 1. 한 쪽의 조상을 먼저 다 저장해두고, 다른 쪽 조상을 탐색하며 공통 조상이 보이는 즉시 반환하는 방법
* 2. 번갈아가며 저장하면서, 공통 조상을 찾는 방법. 1,2 둘 중 뭐가 좋을까. 둘 다 안좋은가 ?

## 코드

```python
# 메모리 초과 # 2번 방법

T = int(input())
N = int(input())
parent = [0] * (10001)
for _ in range(T):
    for _ in range(N-1):
        p, c = map(int, input().split()) # parent[chiled] = parent
        parent[c] = p

    t1, t2 = map(int, input().split())

    ancestor = [t1, t2]
    while True:
        p1, p2 = parent[t1], parent[t2]

        if p1 != 0 and p1 in ancestor:
            print(p1)
            break

        if p2 != 0 and p2 in ancestor:
            print(p2)
            break

        ancestor.append(p1)
        ancestor.append(p2)

        t1 = p1
        t2 = p2
```

```python
# 1번 방법

T = int(input())

for _ in range(T):
    N = int(input())
    parent = [0] * (N + 1)

    for _ in range(N - 1):
        p, c = map(int, input().split())
        parent[c] = p

    t1, t2 = map(int, input().split())

    ancestors = set()

    while t1:
        ancestors.add(t1)
        t1 = parent[t1]

    while t2:
        if t2 in ancestors:
            print(t2)
            break
        t2 = parent[t2]
```
