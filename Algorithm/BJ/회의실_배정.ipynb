## 코드

```python
N = int(input())
arr = []
for _ in range(N):
    arr.append(list(map(int,input().split())))

arr.sort(key=lambda x:(x[1],x[0]))
cur_s, cur_e = -1, -1
result = 0
# result_list = []
for s, e in arr:
    if s >= cur_e:
        # cur_s = s
        cur_e = e
        result += 1
        # result_list.append([s,e])
print(result)



```
