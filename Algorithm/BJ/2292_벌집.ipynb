입력으로 주어진 방까지 최소 개수의 방을 지나서 갈 때 몇 개의 방을 지나는지 출력한다.

## 구상

- 중앙의 방 1 벌집을 중심으로 둘러싼 벌집의 수가 거리별로 얼만큼씩 증가하는지를 구한다. (즉 벌집 테두리의 증가량을 구한다)
- 주어진 N이 속하는 벌집 테두리가 몇번째 인지 구한다

## 코드

```python
1, 2~7, 8~19, 20~37, 38~61
1, 6, 12, 18, 24
  2*3, 4*3, 6*3, 8*3
```

```python
def beehive(n):
    x = 1

    if n == 1:
        return 1

    while True:
        if n > 3*x*(x-1) + 1 and n <= 3*(x+1)*x + 1:
            break
        x += 1
    return x+1
```

```python
beehive()
```
