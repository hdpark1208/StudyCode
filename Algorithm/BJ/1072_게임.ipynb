## 코드

```python
#1 바닐라 > 시간초과
X, Y = map(int,input().split())
Z = Y*100//X
if Z==100:
    print(-1)
else:
    for i in range(1,1000000000):
        Z1 = (Y+i)*100//(X+i)
        if Z!=Z1:
            print(i)
            break
```

```python
#2 이분탐색
X, Y = map(int,input().split())
Z = Y*100//X
if Z>=99:    # 99%일 때도 100%로 올라갈 수 없으므로 이렇게해야함 (기존에 100일때만 신경썼음)
    print(-1)
else:
    start = 1
    end = X    
    answer = 0
    while start<=end:
        mid = (start+end)//2
        Z1 = (Y+mid)*100//(X+mid)
        if Z!=Z1:
            answer = mid
            end = mid-1
        else:
            start = mid+1
    print(answer)
```
