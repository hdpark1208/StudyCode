## 구상

- 라운드가 올라가면서 아긴 참가자는 자신의 번호(A)가 홀수였다면 A//2+1, 짝수였다면 A//2가 된다
- 총 인원(N)도 마찬가지로 홀수였다면 N//2+1, 짝수였다면 N//2가 된다
- 이 과정에서 참가자 A,B가 같은 번호를 부여받는다면 해당 라운드에서 붙게된다
- N 은 고려하지않아도 되고, 무조건 붙게 되므로 -1 출력할 일도 없다

## 코드

```python
N,A,B = map(int,input().split())
round = 0
while A!=B:    
    round += 1
    if A%2==1:
        A = A//2+1
    else:
        A = A//2
    if B%2==1:
        B = B//2+1
    else:
        B = B//2
round
```
