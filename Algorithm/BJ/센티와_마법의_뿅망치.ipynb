```python
# GPT

import heapq

def solve():
    N, H, T = map(int, input().split())
    heap = [-int(input()) for _ in range(N)]
    heapq.heapify(heap)

    cnt = 0

    for _ in range(T):
        mx = -heap[0]          # 현재 최대(꺼내지 말고 peek)
        if mx < H:             # 성공 조건: 최대가 H보다 "작다"
            break
        if mx == 1:            # 1은 때려도 그대로라 더 이상 진행 의미 없음
            break

        # 가장 큰 놈을 반으로 줄여서 그 자리에 바로 갱신
        heapq.heapreplace(heap, -(mx // 2))
        cnt += 1

    mx = -heap[0]
    if mx < H:
        print("YES")
        print(cnt)
    else:
        print("NO")
        print(mx)

solve()
```
