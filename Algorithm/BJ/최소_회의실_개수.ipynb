## 구상

- 1. 최소 회의실 개수만 구하면 되므로, 가장 많이 겹치시는 시간 때 필요한 회의실 개수가 정답이 된다.
- 2. 그 시간을 특정하는 건. 비용이 더 들 것 같고,,,시작시간에 따라 정렬한 다음 필요한 회의실 개수구하며. 그 중 최대값을 구해보자
- 손으로 그려보다 알게 됐는데, 미팅 시작시간을 여는 괄호, 닫는 시간을 닫는 괄호라 생각하면, 이때 가장 중첩이 깊은 괄호가 정답이 되는 것 같다

## 코드

```python
N = int(input())
meeting = []
for _ in range(N):
    meeting.append(list(map(int, input().split())))

time = []
for s, e in meeting:
    time.append((s, 1)) # s : 1
    time.append((e, -1)) # e : -1

time.sort()
max_value = 0
result = 0
for t, tp in time:
    if tp == 1:
        max_value += 1
    else:
        max_value -= 1

    result = max(result, max_value)

print(result)
```
