```python
from collections import deque

n = int(input())
graph = []
for _ in range(n):
    graph.append(list(map(int, input())))

INF = float('inf')
dist = [[INF for _ in range(n)] for _ in range(n)]
dist[0][0] = 0
dq = deque([(0,0)])
dirs = [(-1, 0), (1, 0), (0, -1), (0, 1)]

while dq:
    x, y = dq.popleft()

    for dx, dy in dirs:
        nx = x + dx
        ny = y + dy


        if 0 <= nx < n and 0 <= ny < n:
            nd = dist[nx][ny]

            if graph[nx][ny] == 1: # 흰방
                if dist[x][y] < nd:
                    dist[nx][ny] = dist[x][y]
                    dq.appendleft((nx, ny))

            else: # 검은방
                if dist[x][y] + 1 < nd:
                    dist[nx][ny] = dist[x][y] + 1
                    dq.append((nx, ny))

print(dist[n-1][n-1])
```
