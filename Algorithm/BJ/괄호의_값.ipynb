## 구상

- 일단 스택으로 접근
- 중첩되어 있는 수를 나타내는 변수 관리 (nest)
- 값의 중복이 문제였는데. (), [] 같이 붙어있는 애들만 값을 올려주면 된다

## 코드

```python
from collections import deque
T = input()
result = 0
nest = 1
stack = deque([])
flag = True

for i, p in enumerate(T):
    if p == '(':
        stack.append(p)
        nest *= 2
    elif p == '[':
        stack.append(p)
        nest *= 3

    else:
        if not stack:
            print('0')
            flag = False
            break

        else:
            pre = stack.pop()

            if pre == '(' and p == ')':
                if T[i-1] == '(':
                    result += nest
                nest //= 2

            elif pre == '[' and p == ']':
                if T[i-1] == '[':
                    result += nest
                nest //= 3

            else:
                print('0')
                flag = False
                break

if stack and flag:
    flag = False
    print('0')

if flag:
    print(result)





```

```python
# GPT (리팩토링만) (기대했는데, 굳이? 싶네)
from collections import deque

T = input().strip()

result = 0
nest = 1
stack = deque()

def invalid():
    print(0)
    return None

for i, ch in enumerate(T):
    if ch == '(':
        stack.append('(')
        nest *= 2
    elif ch == '[':
        stack.append('[')
        nest *= 3
    elif ch == ')':
        if not stack or stack.pop() != '(':
            invalid()
            break
        if T[i - 1] == '(':
            result += nest
        nest //= 2
    elif ch == ']':
        if not stack or stack.pop() != '[':
            invalid()
            break
        if T[i - 1] == '[':
            result += nest
        nest //= 3
else:
    # for문이 break 없이 정상 종료된 경우만 여기로 옴
    if stack:
        invalid()
    else:
        print(result)
```
