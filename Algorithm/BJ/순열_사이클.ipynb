```python
def dfs(x):
    visited[x] = True
    next_node = perm[x]
    if not visited[next_node]:
        dfs(next_node)

T = int(input())
for _ in range(T):
    n = int(input())
    temp = list(map(int, input().split()))

    perm = [0] + temp
    visited = [False] * (n + 1)

    count = 0
    for i in range(1, n+1):
        if not visited[i]:
            dfs(i)
            count += 1

    print(count)
```
