```python
import heapq

N, M = map(int, input().split())
times = list(map(int, input().split()))

times.sort(reverse=True)  # 긴 것부터

# 콘센트 M개의 "누적 충전 시간"을 관리하는 최소힙
heap = [0] * M
heapq.heapify(heap)

for t in times:
    x = heapq.heappop(heap)      # 가장 누적시간이 작은 콘센트(가장 빨리 비는 콘센트)
    heapq.heappush(heap, x + t)  # 그 콘센트에 t짜리 기기를 추가

print(max(heap))  # 모든 기기가 끝나는 시간 = 누적시간들 중 최대
```

### 메모



* 충전 시간이 긴 기기부터 꺼내서, 현재까지 누적 충전 시간이 가장 작은 콘센트에 꽂아가며 “마지막으로 끝나는 시간(최대 누적값)”을 최소화하는 문제
