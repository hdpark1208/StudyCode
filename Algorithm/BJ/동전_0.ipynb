```python
from collections import deque

N, K = map(int,input().split())
coins = []
for _ in range(N):
    coins.append(int(input()))

queue = deque(coins)
result = 0

while K > 0:
    # 스택
    coin = queue.pop()
    q = K // coin
    r = K % coin
    K = r

    result += q

print(result)
```
