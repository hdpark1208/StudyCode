## 코드

```python
from collections import deque
import sys

input = sys.stdin.readline

n, l, r = map(int, input().split())
board = [list(map(int, input().split())) for _ in range(n)]

# 상하좌우
dr = (-1, 1, 0, 0)
dc = (0, 0, -1, 1)


def bfs(sr, sc, visited):
    q = deque([(sr, sc)])
    visited[sr][sc] = True

    union = [(sr, sc)]
    total = board[sr][sc]

    while q:
        r0, c0 = q.popleft()

        for k in range(4):
            nr = r0 + dr[k]
            nc = c0 + dc[k]

            if not (0 <= nr < n and 0 <= nc < n):
                continue
            if visited[nr][nc]:
                continue

            if l <= abs(board[r0][c0] - board[nr][nc]) <= r:
                visited[nr][nc] = True
                q.append((nr, nc))
                union.append((nr, nc))
                total += board[nr][nc]

    return union, total


days = 0

while True:
    visited = [[False] * n for _ in range(n)]
    moved = False

    for i in range(n):
        for j in range(n):
            if visited[i][j]:
                continue

            union, total = bfs(i, j, visited)

            if len(union) > 1:
                moved = True
                new_pop = total // len(union)
                for x, y in union:
                    board[x][y] = new_pop

    if not moved:
        break

    days += 1

print(days)
```
