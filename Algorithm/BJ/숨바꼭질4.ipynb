## 구상

- 이동 경로를 추가로 출력해줘야함

## 코드

```python
# GPT # 이전 경로만 저장!

from collections import deque

N, K = map(int, input().split())
MAX = 100000

dist = [-1] * (MAX + 1)
prev = [-1] * (MAX + 1)

q = deque([N])
dist[N] = 0

while q:
    x = q.popleft()

    if x == K:
        break

    for nx in (x - 1, x + 1, x * 2):
        if 0 <= nx <= MAX and dist[nx] == -1:
            dist[nx] = dist[x] + 1
            prev[nx] = x
            q.append(nx)

print(dist[K])

path = []
cur = K
while cur != -1:
    path.append(cur)
    cur = prev[cur]

print(*path[::-1])
```

```python
# 메모리 초과 # 전체 경로 저장

from collections import deque

N, K = map(int, input().split())
MAX = 100000

dist = [-1] * (MAX + 1)
dist[N] = 0

route = [[] for _ in range(MAX + 1)]
route[N].append(N)

q = deque([N])

while q:
    x = q.popleft()
    if x == K:
        break

    for nx in (x - 1, x + 1, x * 2):
        if 0 <= nx <= MAX and dist[nx] == -1:
            dist[nx] = dist[x] + 1
            q.append(nx)
            route[nx] = route[x][:]
            route[nx].append(nx)

print(dist[K])
print(*route[K])
```
