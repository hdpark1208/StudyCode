## 구상

- DP로 접근
- 길이가 홀수인 괄호 문자열은 존재 x



## 코드

```python
MOD = 1_000_000_007

def solve():
    T = int(input().strip())
    Ls = [int(input().strip()) for _ in range(T)]

    maxL = max(Ls)
    maxN = maxL // 2  # 최대 쌍 개수

    # dp[n] = n쌍으로 만들 수 있는 올바른 괄호 문자열 개수
    dp = [0] * (maxN + 1)
    dp[0] = 1

    for n in range(1, maxN + 1):
        s = 0
        # dp[n] = sum_{k=0..n-1} dp[k] * dp[n-1-k]
        for k in range(n):
            s = (s + dp[k] * dp[n - 1 - k]) % MOD
        dp[n] = s

    for L in Ls:
        if L % 2 == 1:
            print(0)
        else:
            print(dp[L // 2])

solve()
```
