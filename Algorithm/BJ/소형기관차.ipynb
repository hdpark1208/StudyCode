```python
def solve():
    N = int(input().strip())
    a = list(map(int, input().split()))
    M = int(input().strip())

    # 1-indexed prefix sum
    ps = [0] * (N + 1)
    for i in range(1, N + 1):
        ps[i] = ps[i - 1] + a[i - 1]

    # dp[i][j] : using i trains within first j cars (i=0..3)
    dp = [[0] * (N + 1) for _ in range(4)]

    for i in range(1, 4):  # 1..3 trains
        start = i * M
        for j in range(start, N + 1):
            take = dp[i - 1][j - M] + (ps[j] - ps[j - M])
            dp[i][j] = max(dp[i][j - 1], take)

    print(dp[3][N])

solve()
```
