## 구상

- 구상1. 전 단계의 가장 빠른 시간이, (정답 시간 - 1)이 되면 되려나? 이 문제의 경우 (x-1, x+1, 2*x)의 단계가 있으므로 최대 3 ? 일리가 없는데...
- 구상2. 방문 조건을 없애고, 값을 다 받아들인 후, 최소값들의 개수 구하기

## 코드

```python
from collections import deque

N, K = map(int, input().split())
MAX = 100000

dist = [-1] * (MAX + 1)
cnt = [0] * (MAX + 1)

q = deque([N])
dist[N] = 0
cnt[N] = 1

while q:
    x = q.popleft()

    for nx in (x - 1, x + 1, x * 2):
        if 0 <= nx <= MAX:
            # 처음 방문
            if dist[nx] == -1:
                dist[nx] = dist[x] + 1
                cnt[nx] = cnt[x]
                q.append(nx)

            # 이미 방문했지만, 같은 최단 거리로 또 도달
            elif dist[nx] == dist[x] + 1:
                cnt[nx] += cnt[x]

print(dist[K])
print(cnt[K])
```
