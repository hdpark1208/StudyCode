## 구상

* 모든 두 쌍 사이의 최소값을 구해놓고, 입력에 맞춰서 만족하는 개수 출력

## 코드

```python
from collections import deque
import sys

input = sys.stdin.readline

N, Q = map(int, input().split())

graph = [[] for _ in range(N + 1)]

for _ in range(N - 1):
    p, q, r = map(int, input().split())
    graph[p].append((q, r))
    graph[q].append((p, r))


def bfs(k, start):
    visited = [False] * (N + 1)
    visited[start] = True
    q = deque([start])
    cnt = 0

    while q:
        cur = q.popleft()

        for nxt, usado in graph[cur]:
            if not visited[nxt] and usado >= k:
                visited[nxt] = True
                q.append(nxt)
                cnt += 1

    return cnt


for _ in range(Q):
    k, v = map(int, input().split())
    print(bfs(k, v))
```
