```python
# BFS
from collections import deque

def solve():
    n = int(input())
    adj = [[] for _ in range(n + 1)]
    for _ in range(n - 1):
        a, b = map(int, input().split())
        adj[a].append(b)
        adj[b].append(a)

    parent = [0] * (n + 1)
    parent[1] = -1  # 루트 표시

    q = deque([1])
    while q:
        cur = q.popleft()
        for nxt in adj[cur]:
            if parent[nxt] == 0:      # 아직 방문 전
                parent[nxt] = cur     # cur이 nxt의 부모
                q.append(nxt)

    for i in range(2, n + 1):
        print(parent[i])

solve()
```

```python
# DFS
import sys
sys.setrecursionlimit(10**7)

def solve():
    n = int(input())
    adj = [[] for _ in range(n + 1)]

    for _ in range(n - 1):
        a, b = map(int, input().split())
        adj[a].append(b)
        adj[b].append(a)

    parent = [0] * (n + 1)

    def dfs(cur):
        for nxt in adj[cur]:
            if parent[nxt] == 0:     # 아직 방문 안 함
                parent[nxt] = cur    # cur이 nxt의 부모
                dfs(nxt)

    parent[1] = -1   # 루트 표시
    dfs(1)

    for i in range(2, n + 1):
        print(parent[i])

solve()
```
