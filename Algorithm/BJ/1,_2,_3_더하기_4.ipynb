## 코드

```python
del max
```

```python
T = int(input())
arr = [ int(input()) for i in range(T)]
max_num = max(arr)
dp = [0] * (max_num + 1)
dp[0] = 1
for n in (1,2,3):
    for s in range(n, max_num + 1):
        dp[s] += dp[s - n]

for i in arr:
    print(dp[i])


```
