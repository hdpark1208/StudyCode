```python
from collections import deque

def move(idx, direction):
    if direction == 1:      # 시계
        wheel[idx].appendleft(wheel[idx].pop())
    else:                   # 반시계
        wheel[idx].append(wheel[idx].popleft())

wheel = [deque(input().strip()) for _ in range(4)]

K = int(input())
commands = [tuple(map(int, input().split())) for _ in range(K)]

for num, direction in commands:
    idx = num - 1
    rot = [0] * 4
    rot[idx] = direction

    # 왼쪽 전파
    for i in range(idx, 0, -1):
        if wheel[i][6] != wheel[i - 1][2]:
            rot[i - 1] = -rot[i]
        else:
            break

    # 오른쪽 전파
    for i in range(idx, 3):
        if wheel[i][2] != wheel[i + 1][6]:
            rot[i + 1] = -rot[i]
        else:
            break

    # 실제 회전
    for i in range(4):
        if rot[i] != 0:
            move(i, rot[i])

score = 0
for i in range(4):
    if wheel[i][0] == '1':
        score += (1 << i)

print(score)
```

```python
# 지저분 # 헷갈림

def move(num, rotation):
    if rotation == 1:
        x = wheel[num].pop()
        wheel[num].appendleft(x)

    else:
        x = wheel[num].popleft()
        wheel[num].append(x)

from collections import deque
wheel = []
for _ in range(4):
    wheel.append((deque(list(input()))))

K = int(input())
command = []
for _ in range(K):
    command.append(map(int,input().split()))

# 맞닿는 곳 인덱스 : 2(우), 6(좌) # num : 1 ~ 4

for num, r in command:
    right = wheel[num - 1][2]
    left = wheel[num - 1][6]
    move(num - 1, r)
    # 오른쪽 방향
    i = 1

    while num + i < 4:
        touch = i % 2

        if wheel[num + i - 1][6] == right:
            break
        else:
            right = wheel[num + i - 1][2]
            i += 1
            move(num + i - 1, r * (-1) ** touch)

    # 왼쪽 방향
    j = 1

    while num - j > 0:
        touch = j % 2

        if wheel[num - j - 1][2] == left:
            break
        else:
            left = wheel[num - j - 1][6]
            j += 1
            move(num - j - 1, r * (-1) ** touch)

    print('wheel:',wheel)

```
