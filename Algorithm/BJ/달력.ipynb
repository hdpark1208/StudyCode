## 구상

- 주어진 입력을 규칙에 따라 배열한 그래프를 만든다
    - 이게 생각보다 어렵네...
- dfs를 활용하여 해당 그래프에서 이어진(연속된 일정) 곳 들을 구한다
    - 기각. 그래프만 잘 만들면 필요도 없을 것 같다
- 각 연속된 일정들의 x,y 최대, 최소 (직사각형) 값을 구하고 면적을 구하여 더한다


## 코드

```python
N = int(input())
calendar = []
for _ in range(N):
    calendar.append(list(map(int, input().split())))

calendar.sort(key= lambda x: (x[0],-x[1]))

visited = [False] * 366
count_arr = [0] * 366

for s, e in calendar:
    for i in range(s, e + 1):
        if visited[i] == False:
            visited[i] = True
            count_arr[i] += 1
        else:
            count_arr[i] += 1

result = 0
v = 0
l = 0

for i in range(366):
    if visited[i] == True:
        v = max(v, count_arr[i])
        l += 1

        if i == 365:
            result += v * l

    elif visited[i] == False:
        result += v * l
        v = 0
        l = 0



print(result)
```

```python
# GPT 리펙토링 코드(답은 틀림 ㅋㅋ) # result += width * max_height 를 추가 or 더미 추가로 해결

N = int(input())
count_arr = [0] * 366

for _ in range(N):
    s, e = map(int, input().split())
    for i in range(s, e + 1):
        count_arr[i] += 1

result = 0
width = 0
max_height = 0

for i in range(366):
    if count_arr[i] > 0:
        width += 1
        max_height = max(max_height, count_arr[i])
    else:
        result += width * max_height
        width = 0
        max_height = 0

# 수정된 부분
result += width * max_height

print(result)
```
