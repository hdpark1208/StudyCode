## 코드

```python
N = int(input())
board = [list(map(int, input().split())) for _ in range(N)]

dp = [[0] * N for _ in range(N)]
dp[0][0] = 1

for r in range(N):
    for c in range(N):
        if dp[r][c] == 0:
            continue
        if r == N - 1 and c == N - 1:
            break

        k = board[r][c]
        nr = r + k
        nc = c + k

        if nr < N:
            dp[nr][c] += dp[r][c]
        if nc < N:
            dp[r][nc] += dp[r][c]

print(dp[N-1][N-1])
```
