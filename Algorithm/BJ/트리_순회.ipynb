## 코드

```python
def preorder(node):
    if node == '.':
        return
    left, right = tree[node]
    pre.append(node)
    preorder(left)
    preorder(right)

def inorder(node):
    if node == '.':
        return
    left, right = tree[node]
    inorder(left)
    ino.append(node)
    inorder(right)

def postorder(node):
    if node == '.':
        return
    left, right = tree[node]
    postorder(left)
    postorder(right)
    post.append(node)


N = int(input())
tree = {}

for _ in range(N):
    a, b, c = input().split()
    tree[a] = (b, c)

pre, ino, post = [], [], []

preorder('A')
inorder('A')
postorder('A')

print(''.join(pre))
print(''.join(ino))
print(''.join(post))
```

| 이름             | 의미                    |
| -------------- | --------------------- |
| **pre-order**  | 루트를 **먼저(pre)** 방문    |
| **in-order**   | 루트를 **사이(in)** 에 방문   |
| **post-order** | 루트를 **나중(post)** 에 방문 |
