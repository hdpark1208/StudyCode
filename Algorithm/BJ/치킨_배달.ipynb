## 구상

- 일단 거리 계산 함수
- 백트래킹 ? 조합 ? 조합!
- X개의 치킨집이 주어졌다면 그 중 M개의 치킨집을 조합을 통해 구하고,
- 해당 집 마다 가장 가까운 치킨집을 다 찾은 다음 거리를 구해서, 그 거리를 다 합친다.


## 코드

```python
def dist(chickens, houses): # chickens = ((1, 2), (4, 1)) # houses = [[1,4],[2,1],[4,4],[4,5]]
    result = 0
    for h in houses:
        dist = float('inf')
        for c in chickens:
            dist = min(dist, abs(h[0] - c[0]) + abs(h[1] - c[1]))

        result += dist

    return result

from itertools import combinations

N, M = map(int,input().split())
graph = []
# answer_list = []
answer = float('inf')

for _ in range(N):
    graph.append(list(map(int, input().split())))

H = []
C = []

# C_num = 0

for i in range(N):
    for j in range(N):
        if graph[i][j] == 1:
            H.append((i+1, j+1))
        elif graph[i][j] == 2:
            C.append((i+1, j+1))
            # C_num += 1

for c in combinations(C, M):
    answer = min(answer, dist(c, H))

print(answer)
```
