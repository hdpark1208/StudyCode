```python
# 1
def solve():
    T = int(input())
    queries = [int(input()) for _ in range(T)]
    max_n = max(queries) if queries else 0
    print(queries)
    print(max_n)

    # count0[n], count1[n]
    count0 = [0] * (max_n + 1)
    count1 = [0] * (max_n + 1)

    # base cases
    count0[0] = 1
    count1[0] = 0
    if max_n >= 1:
        count0[1] = 0
        count1[1] = 1

    # dp
    for n in range(2, max_n + 1):
        count0[n] = count0[n - 1] + count0[n - 2]
        count1[n] = count1[n - 1] + count1[n - 2]

    for n in queries:
        print(count0[n], count1[n])

solve()
```

```python
# 2
def solve():
    n = int(input())

    if n == 0:
        print(0)
        print(0)
        return None

    fibo = [0] * (abs(n) + 3)

    # base cases
    fibo[1] = 1

    # dp
    if n > 0:
        for n in range(2, n + 1):
            fibo[n] = (fibo[n-1] + fibo[n-2])
        print(1)
        print(fibo[n]% 1000000000)
    else:
        fibo[2] = -1

        for i in range(1, -n + 1): # n : 0, -1, -2, ..., -n
            fibo[i + 2] = abs(fibo[i] - fibo[i + 1]) # % 1000000000
        if fibo[-n] > 0:
            print(1)
        else:
            print(-1)

        print(abs(fibo[-n]) % 1000000000)
        print(fibo)

solve()
```
