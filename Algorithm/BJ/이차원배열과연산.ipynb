## 코드

```python
def operator(line):
    cnt_list = []
    for i in set(line):
        if i != 0:
            cnt_list.append((i, line.count(i)))
    cnt_list.sort(key = lambda x: (x[1],x[0]))

    # for pair in cnt_list:
    #     for x in pair:
    #         result.append(x)
    result = [x for pair in cnt_list for x in pair]

    return result

def zero_padding(matrix, max_l):
    for i, line in enumerate(matrix):
        l = len(line)

        if l < max_l:
            size = max_l - l
            matrix[i] += [0] * size

    return matrix

def cuting(matrix, l = 100):
    for i, line in enumerate(matrix):
        matrix[i] = matrix[i][:l]

    return matrix

r, c, k = map(int,input().split())
A = []
for _ in range(3):
    A.append(list(map(int, input().split())))

w = len(A[0]) # 열의 개수
h = len(A) # 행의 개수
t = 0
answer = False
while t <= 100:

    if r <= h and c <= w:
        if A[r-1][c-1] == k:
            print(t)
            answer = True
            break

    if w <= h:
        new_A = []
        max_l = 0
        for line in A:
            flatted_line = operator(line)
            new_A.append(flatted_line)

            max_l = max(max_l, len(flatted_line))

        A = zero_padding(new_A, max_l)
        if max_l > 100:
            A = cuting(A)

    else:
        new_A = []
        A_t = []
        max_l = 0

        for i in zip(*A):
            A_t.append(i)


        for line in A_t:
            flatted_line = operator(line)
            new_A.append(flatted_line)

            max_l = max(max_l, len(flatted_line))



        A_t = zero_padding(new_A, max_l)
        if max_l > 100:
            A_t = cuting(A_t)

        A = []
        for i in zip(*A_t):
            A.append(i)

    w = len(A[0])
    h = len(A)
    t += 1

if not answer:
    print(-1)
```
