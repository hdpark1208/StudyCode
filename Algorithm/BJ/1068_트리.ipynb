## 구상

- 입력으로 0번 노드 부터 N-1번 노드의 부모 노드가 주어지므로  
index는 자신의 노드 번호, 값은 부모 노드를 가지는 리스트를 만든다
-  삭제할 노드가 주어지면 해당 노드와 그 자식들을 삭제 처리시키는 dfs를 구현한다

## 코드

```python
def dfs(target_node,arr):
    arr[target_node] = -2    # 해당 노드 삭제 처리
    for i in range(len(arr)):    # 리스트 전체를 돌면서
        if arr[i]==target_node:  # 어떤 노드가 target_node를 부모로 가진다면
            dfs(i,arr)               # 해당 노드에 대해 dfs 처리

def leaf_count(arr):
    child = [False]*len(arr)    # 자식 유무 판정을 위한 리스트
    for i in range(len(arr)):
        if arr[i]!=-2 and arr[i]!= -1:    # 삭제된 노드, 부모 노드가 아니라면
            child[arr[i]] = True    # 해당 노드의 값을 자식 있음(True)으로 변경
        if arr[i] == -2:    # 삭제된 노드라면
            child[i] = 'X'    #  해당 노드의 값을 'X'로 표시
    return child.count(False)    # 자식이 없는 노드 수 출력 (리프 노드)

N = int(input())
arr = list(map(int,input().split()))
target = int(input())

dfs(target,arr)
print(leaf_count(arr))
```
