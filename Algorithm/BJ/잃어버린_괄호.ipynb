## 코드

```python
# 단순하게 풀었다
expression = input()
arr = []
temp = ''

for i in expression:
    if i == '+' or i == '-':
        arr.append(int(temp))
        arr.append(i)
        temp = ''

    else:
        temp += i

arr.append(int(temp))

result = 0
idx = 50
for i in range(len(arr)):
    if arr[i] == '-':
        idx = i
        break

for i in range(len(arr)):
    if i < idx and type(arr[i]) == int:
        result += arr[i]

    elif i > idx and type(arr[i]) == int:
        result -= arr[i]

print(result)
```
