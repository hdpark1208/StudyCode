## 코드

```python
def MVF(matrix):
    result = 0
    for i in range(N):
        friends = [False]*N
        for j in range(N):
            if matrix[i][j]=='Y':      # 친구이면
                friends[j] = True    # 친구배열에 저장
                for k in range(N):
                    if matrix[j][k]=='Y':
                        friends[k] = True
        friends[i] = False    # 자신이 포함되는것 방지
        count = friends.count(True)
        result = max(result,count)
    return result

N = int(input())
matrix = []
for _ in range(N):
    matrix.append(list(input()))

MVF(matrix)
```
