```python
# import sys
# sys.setrecursionlimit(10**7)
# input = sys.stdin.readline().rstrip()

N, M = map(int, input().split())

# 인접 리스트
graph = [[] for _ in range(N + 1)]
for _ in range(N - 1):
    a, b, d = map(int, input().split())
    graph[a].append((b, d))
    graph[b].append((a, d))

def dfs(cur, target, dist):
    if cur == target:
        return dist

    for nxt, w in graph[cur]:
        if not visited[nxt]:
            visited[nxt] = True
            res = dfs(nxt, target, dist + w)
            if res != -1:
                return res
    return -1

for _ in range(M):
    u, v = map(int, input().split())
    visited = [False] * (N + 1)
    visited[u] = True
    print(dfs(u, v, 0))
```
