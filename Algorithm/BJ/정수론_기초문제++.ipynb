```python
# 4134
import math
def is_prime(n):
    while True:
        for i in range(2,int(math.sqrt(n))+1):
            if n % i == 0:
                return False

        return True

def least_prime(n):
    if n == 1 or n == 0:
        return 2
    while True:
        if is_prime(n):
            return n
        else:
            n += 1
n = int(input())
for _ in range(n):
    print(least_prime(int(input())))
```

```python
# 17103
import math
import sys

def sieve(limit):
    is_prime = [True] * (limit + 1)
    is_prime[0] = is_prime[1] = False

    for i in range(2, int(math.sqrt(limit)) + 1):
        if is_prime[i]:
            for j in range(i*i, limit +1, i):
                is_prime[j] = False

    return is_prime

def Goldbach(n,sieve):
    if n == 4:
        return 1
    count = 0
    for i in range(1, int(n/2)+1, 2):
        if sieve[i] and sieve[n-i]:
            count += 1

    return count

limit = 1000000
sieve_1 = sieve(limit)
T = int(input())

for _ in range(T):
    print(Goldbach(int(input()), sieve_1))
```
