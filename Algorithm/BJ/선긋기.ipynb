```python
# GPT 도움

N = int(input())
lines = [tuple(map(int, input().split())) for _ in range(N)]

lines.sort()  # 시작점 기준 정렬, 시작점 같으면 끝점 기준 정렬

cur_s, cur_e = lines[0]
answer = 0

for s, e in lines[1:]:
    if s <= cur_e:  # 겹치거나 붙어 있는 경우
        cur_e = max(cur_e, e)
    else:  # 안 겹치면 이전 구간 길이 확정
        answer += cur_e - cur_s
        cur_s, cur_e = s, e

answer += cur_e - cur_s  # 마지막 구간 더하기

print(answer)
```

```python
# 메모리 초과

N = int(input())
start = []
end = []

for _ in range(N):
    s, e = map(int, input().split())
    start.append(s)
    end.append(e)

MIN = min(start)
MAX = max(end)
L = MAX - MIN

visited = [False] * L

for s, e in zip(start, end):
    visited[s:e] = [True] * (e - s)

print(visited.count(True))
```
