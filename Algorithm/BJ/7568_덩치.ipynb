```python
# 실패. 조잡

import sys
# input = sys.stdin.readline

N = int(input())
size = dict()
height_list = []
# weight_list = []
for i in range(1,N+1):

    data = tuple(map(int, input().split()))
    size[i] = data
    height_list.append(data)

height_list.sort(reverse=True)
weight_list = height_list.sort(reverse=True, key=lambda x:x[1])

rank = 1
rank_list = []
for h in height_list:
    for w in weight_list:
        if h[0] > w[0] and h[1] > w[1] and h not in rank_list:
            rank_list.append((h,rank))




```

```python
# GPT

N = int(input())
people = [tuple(map(int, input().split())) for _ in range(N)]

ranks = []

for i in range(N):
    rank = 1  # 기본 등수는 1
    for j in range(N):
        if i != j:
            if people[j][0] > people[i][0] and people[j][1] > people[i][1]:
                rank += 1
    ranks.append(rank)

print(*ranks)
```
