```python
n = int(input())
target = int(input())

arr = [[0]*n for _ in range(n)]

# 방향: 아래, 오른쪽, 위, 왼쪽
# dirs = [(1,0), (0,1), (-1,0), (0,-1)]

# 방향 수정
dirs = [(0,1), (-1,0), (0,-1), (1,0)]

x, y = 0, 0
cur = n * n
arr[x][y] = cur

# target 위치 저장
tx, ty = x, y

step = n - 1
d = 0

while step > 0:
    if step == n - 1:
        dx, dy = dirs[3]
        for _ in range(step):

            x += dx
            y += dy
            cur -= 1
            arr[x][y] = cur
            if cur == target:
                tx, ty = x, y

    for _ in range(2):  # 같은 step을 두 번 사용
        dx, dy = dirs[d % 4]
        for _ in range(step):
            x += dx
            y += dy
            cur -= 1
            arr[x][y] = cur
            if cur == target:
                tx, ty = x, y
        d += 1
    step -= 1

for row in arr:
    print(*row)

# 문제는 1-indexed
print(tx+1, ty+1)
```
