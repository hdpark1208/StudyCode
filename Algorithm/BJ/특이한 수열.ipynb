## 구상

* 연합이 2개 이상인 경우 어떻게 처리하지

## 코드

```python
# GPT 수정본

n, k = map(int, input().split())

if k == n:
    print("Impossible")
elif k == 0:
    ans = list(range(2, n + 1)) + [1]
    print(*ans)
elif k == 1:
    if n == 1:
        print("Impossible")
    elif n == 2:
        print(1, 2)
    else:
        ans = [n, 2, 1] + list(range(3, n))
        print(*ans)
else:
    m = n - k
    ans = list(range(1, n + 1))
    ans[:m] = ans[1:m] + [1]
    print(*ans)
```

```python
# GPT 초기 코드 # k == 1 에서 반례 발견 but 제출 통과

n, k = map(int, input().split())

if k == 1 or k == n:
    print("Impossible")
else:
    m = n - k
    ans = list(range(1, n + 1))

    # 앞의 m개를 한 칸 왼쪽 회전
    # 1,2,3,...,m -> 2,3,4,...,m,1
    ans[:m] = ans[1:m] + [1]

    print(*ans)
```

맞습니다.
네 예시가 반례고, 내가 전에 k=1은 불가능이라고 한 건 틀렸습니다.

직접 확인해보면:

순열: [4, 2, 1, 3]
위치별 gcd:
gcd(1, 4) = 1
gcd(2, 2) = 2 ← 성공
gcd(3, 1) = 1
gcd(4, 3) = 1

따라서 성공하는 위치가 정확히 1개이므로 k=1 가능입니다.

내가 앞서 말한 부분은 정정해야 합니다.
