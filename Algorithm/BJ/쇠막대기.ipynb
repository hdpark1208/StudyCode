## 코드

```python
def solve():
    s = input().strip()

    stack = []
    ans = 0

    for i, ch in enumerate(s):
        if ch == '(':
            stack.append('(')
        else:  # ch == ')'
            stack.pop()
            if s[i - 1] == '(':      # 레이저 "()"
                ans += len(stack)
            else:                     # 막대기 끝
                ans += 1

    print(ans)

solve()
```
