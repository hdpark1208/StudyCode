## 코드

```python
# 시간 초과

import copy
def rotate_matrix(matrix, query):
    answer = copy.deepcopy(matrix)
    value = 9999
    x1, y1, x2, y2 = query[0] - 1, query[1] - 1, query[2] - 1, query[3] - 1

    # 상변
    for i in range(y1, y2):
        if matrix[x1][i] < value:
            value = matrix[x1][i]


        if i == y1:
            answer[x1][i] = matrix[x1 + 1][i]
        else:
            answer[x1][i] = matrix[x1][i -1]

    # 우변
    for i in range(x1, x2):
        if matrix[i][y2] < value:
            value = matrix[i][y2]

        if i == x1:
            answer[i][y2] = matrix[i][y2 - 1]
        else:
            answer[i][y2] = matrix[i - 1][y2]

    # 하변
    for i in range(y2, y1, -1):
        if matrix[x2][i] < value:
            value = matrix[x2][i]

        if i == y2:
            answer[x2][i] = matrix[x2 - 1][i]
        else:
            answer[x2][i] = matrix[x2][i + 1]

    # 좌변
    for i in range(x2, x1, -1):
        if matrix[i][y1] < value:
            value = matrix[i][y1]

        if i == x2:
            answer[i][y1] = matrix[i][y1 + 1]
        else:
            answer[i][y1] = matrix[i + 1][y1]

    return answer, value



def solution(rows, columns, queries):
    matrix = [[i * columns + j + 1 for j in range(columns)] for i in range(rows)]
    answer = []
    for query in queries:
        matrix, value = rotate_matrix(matrix, query)
        answer.append(value)

    return answer
```

```python
rows = 3
cols = 3
queries = [[1,1,2,2],[1,2,2,3],[2,1,3,2],[2,2,3,3]]
```

```python
solution(rows,cols,queries)
```
