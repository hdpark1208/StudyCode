## 구상

## 코드

```python
def solution(left, right):
    answer = 0
    for i in range(left, right + 1):
        if divisor(i) == 'even':
            answer += i
        else:
            answer -= i
    return answer
```

```python
def divisor(n):
    result = [1,n]
    ans = []
    for i in range(2,int(n**(1/2))+1):
        if n % i == 0:
            result.append(i)
            result.append(n//i)
    result = set(result)
    if len(result) % 2 == 0:
        return 'even'
    else:
        return 'odd'
```
