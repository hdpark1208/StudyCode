```python
def solution(lottos, win_nums): # 6개 일치 1등, 0개 일치 6등
    answer = []
    x = 0
    match_cnt = 0

    for i in lottos:
        if i == 0:
            x += 1
        else:
            if i in win_nums:
                match_cnt += 1

    max = 7 - (match_cnt + x)
    min = 7 - (match_cnt)

    if max == 7:
        max = 6

    if min == 7:
        min = 6

    answer.append(max)
    answer.append(min)

    result = [max, min]

    return result
```
