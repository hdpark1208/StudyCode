## 코드

```python
def solution(s):

    trans_cnt = 0
    zero_cnt = 0

    while len(s) > 1:

        temp_cnt = 0

        for i in s:
            if i == '0':
                temp_cnt += 1

        s = binary_trans(len(s) - temp_cnt)
        trans_cnt += 1
        zero_cnt += temp_cnt



    return [trans_cnt, zero_cnt]

def binary_trans(n):
    if n == 0:
        return "0"

    result = ""
    while n > 0:
        remainder = n % 2
        result += str(remainder)
        n = n // 2

    return result[::-1]
```
