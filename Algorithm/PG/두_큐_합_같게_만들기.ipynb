```python
from collections import deque

def solution(queue1, queue2):
    q1 = deque(queue1)
    q2 = deque(queue2)

    sum1 = sum(q1)
    sum2 = sum(q2)
    total = sum1 + sum2

    if total % 2 != 0:
        return -1

    target = total // 2
    cnt = 0
    limit = len(q1) * 3  # 충분한 횟수, 최악의 경우 2n 번 + 여유

    while cnt <= limit:
        if sum1 == target:
            return cnt
        elif sum1 > target:
            x = q1.popleft()
            sum1 -= x
            q2.append(x)
            sum2 += x
        else:
            x = q2.popleft()
            sum2 -= x
            q1.append(x)
            sum1 += x
        cnt += 1

    return -1
```
