# 숫자 문자열과 영단어

image.png

## 입력 출력 및 예제


* 입력

image.png

* 출력

* 예제

image.png

## 구상

- 숫자는 그대로, 문자열은 숫자로 바꿔서 출력해야하는데,
문자열이 연속으로 주어진 경우가 관건이다
- 문자열이 들어올 경우, 모든 경우에 대해 점검해서...해보자 좋은 아이디어가 안떠오른다


## 코드

```python
def solution(s):
    mapping = {
        "zero": "0",
        "one": "1",
        "two": "2",
        "three": "3",
        "four": "4",
        "five": "5",
        "six": "6",
        "seven": "7",
        "eight": "8",
        "nine": "9",
    }

    answer = ""
    temp = ""

    for ch in s:
        if ch.isdigit():
            answer += ch
        else:
            temp += ch
            if temp in mapping:
                answer += mapping[temp]
                temp = ""
    return int(answer)
```

### 메모
- temp 버퍼를 만들고 mapping 딕셔너리를 통해, 해당 문자열이 완전히 매핑될 때 까지 누적시키고, 완성되면 대응되는 숫자를 추가한 뒤 버퍼를 초기화시키는 방법!!
