```python
def h_to_m(hour):
    h = hour // 100
    m = hour % 100
    return 60 * h + m

def solution(schedules, timelogs, startday):
    N = len(schedules)
    answer = N
    for i in range(N):
        day = startday

        for time in timelogs[i]:

            if (h_to_m(time) - h_to_m(schedules[i])) <= 10 or day % 7 == 6 or day % 7 == 0:
                day += 1
            else:
                answer -= 1
                break

    return answer
```
