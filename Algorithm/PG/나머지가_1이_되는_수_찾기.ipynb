```python
def solution(n):
    x = 2
    while True:
        k = n % x
        if k == 1:
            return x
        x += 1
```

```python
def solution(numbers):
    answer = 0
    for i in range(1,10):
        if i not in numbers:
            answer += i

    return answer
```
