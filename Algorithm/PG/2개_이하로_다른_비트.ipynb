## 코드

```python
def f(x: int) -> int:
    # 짝수인 경우
    if x % 2 == 0:
        return x + 1

    # 홀수인 경우: 이진 문자열로 변환한 뒤 앞에 '0'을 붙여 예외 처리
    b = '0' + bin(x)[2:]
    idx = b.rfind('0')  # 오른쪽부터 첫 0의 위치
    # 그 위치를 '1'로, 바로 다음 비트를 '0'으로 바꿈
    b = b[:idx] + '10' + b[idx+2:]
    return int(b, 2)

def solution(numbers):
    return [f(x) for x in numbers]
```
