```python
def solution(numbers, hand): # (0) : 0, (7,8,9) : 1, (4,5,6) : 2, (1,2,3) : 3
    answer = ''


    L_pos = '*'
    R_pos = '#'

    for num in numbers:
        if num in [1,4,7]:
            answer += 'L'
            L_pos = num
        elif num in [3,6,9]:
            answer += 'R'
            R_pos = num
        else:
            L_dist = dist(L_pos, num)
            R_dist = dist(R_pos, num)

            if L_dist < R_dist:
                answer += 'L'
                L_pos = num
            elif R_dist < L_dist:
                answer += 'R'
                R_pos = num
            else:
                if hand == 'left':
                    answer += 'L'
                    L_pos = num
                else:
                    answer += 'R'
                    R_pos = num

    return answer

def dist(a, b):
    pos = {
    1:(0,0), 2:(0,1), 3:(0,2),
    4:(1,0), 5:(1,1), 6:(1,2),
    7:(2,0), 8:(2,1), 9:(2,2),
    '*':(3,0), 0:(3,1), '#':(3,2)
    }
    r1, c1 = pos[a]
    r2, c2 = pos[b]
    return abs(r1 - r2) + abs(c1 - c2)
```

```python
pos = {
    1:(0,0), 2:(0,1), 3:(0,2),
    4:(1,0), 5:(1,1), 6:(1,2),
    7:(2,0), 8:(2,1), 9:(2,2),
    '*':(3,0), 0:(3,1), '#':(3,2)
}
```

```python
numbers = [1, 3, 4, 5, 8, 2, 1, 4, 5, 9, 5]
hand = "right"
```
