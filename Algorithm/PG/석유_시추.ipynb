```python
def label_and_count_components(grid):
    from string import ascii_uppercase

    n, m = len(grid), len(grid[0])
    visited = [[False] * m for _ in range(n)]
    label_grid = [[''] * m for _ in range(n)]
    label_counts = {}
    label_iter = iter(ascii_uppercase)  # A, B, C, ...

    def dfs(x, y, label):
        stack = [(x, y)]
        count = 0
        while stack:
            cx, cy = stack.pop()
            if (0 <= cx < n) and (0 <= cy < m) and not visited[cx][cy] and grid[cx][cy] == 1:
                visited[cx][cy] = True
                label_grid[cx][cy] = label
                count += 1
                # 상하좌우
                stack.extend([
                    (cx-1, cy), (cx+1, cy),
                    (cx, cy-1), (cx, cy+1)
                ])
        return count

    for i in range(n):
        for j in range(m):
            if grid[i][j] == 1 and not visited[i][j]:
                label = str(len(label_counts))  # 라벨을 숫자 문자열로
                count = dfs(i, j, label)
                label_counts[label] = count

    return label_counts, label_grid


def solution(grid):
    n = len(grid)
    m = len(grid[0])

    answer = []
    label, labeled_grid = label_and_count_components(grid)

    for col in range(m):
        total_oil = 0
        col_oil = set([row[col] for row in labeled_grid])
        for i in col_oil:
            if i != '':
                total_oil += label[i]
        answer.append(total_oil)

    return max(answer)
```
