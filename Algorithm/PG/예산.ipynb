## 코드

```python
def solution(d, budget):
    answer = 0
    d.sort()

    for c in d:
        budget -= c

        if budget >= 0:
            answer += 1
        else:
            break

    return answer
```

```python
# d = [1,3,2,5,4]
d = [2,2,3,3]
budget = 10
```

```python
solution(d,budget)
```
