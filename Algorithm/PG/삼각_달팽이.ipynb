```python
def solution(n):
    answer = []

    arr = [[0] * (i+1) for i in range(n)]

    num = 1
    i, j = 0, 0

    while num <= n * (n+1) // 2:
        arr[i][j] = num
        num += 1

        if i + 1 < n and arr[i + 1][j] == 0: # 아래로 갈 수 있으면,,
            i += 1
        elif j + 1 < len(arr[i]) and arr[i][j+1] == 0:
            j += 1

        else:
            i -= 1
            j -= 1

    for row in arr:
        answer.extend(row)


    return answer
```

```python
# gpt
def solution(n):
    arr = [[0] * (i + 1) for i in range(n)]

    # ↓, →, ↖
    dx = [1, 0, -1]
    dy = [0, 1, -1]

    x, y = 0, 0
    d = 0
    num = 1

    for step in range(n, 0, -1):
        for _ in range(step):
            arr[x][y] = num
            num += 1

            if _ == step - 1:
                break

            x += dx[d]
            y += dy[d]

        d = (d + 1) % 3
        x += dx[d]
        y += dy[d]

    answer = []
    for row in arr:
        answer.extend(row)

    return answer
```
