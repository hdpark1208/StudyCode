## 구상

* 각 응시자의 위치에서 BFS를 수행. 조건에 따라 거리두기를 어겼으면 바로 break 후 0, 아니면 1

## 코드

```python
from collections import deque

def dis(x1, y1, x2, y2):
    return abs(x2 - x1) + abs(y2 - y1)

def bfs(x,y,place): # (x,y) 위치의 응시자가 규칙을 지키고 있는지
    dirs = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    visited = [[False] * 5 for _ in range(5)]
    dq = deque([(x,y)])
    visited[x][y] = True
    result = True
    sx, sy = x, y

    while dq:
        x, y = dq.popleft()
        for dx, dy in dirs:
            nx = x + dx
            ny = y + dy

            if dis(sx, sy, nx, ny) <= 2 and 0 <= nx < 5 and 0 <= ny < 5 and not visited[nx][ny]:
                if place[nx][ny] == "P":
                    result = False
                    break

                elif place[nx][ny] == "O":
                    dq.append((nx, ny))
                    visited[nx][ny] = True

                else: # "X"
                    pass


    return result



def solution(places):
    answer = []

    for place in places:
        applicant_list = []
        for x in range(len(place)):
            for y in range(len(place[0])):
                if place[x][y] == "P":
                    applicant_list.append((x,y))

        result = True
        for x,y in applicant_list:
            if not bfs(x, y, place):
                result = False
                break

        if result:
            answer.append(1)
        else:
            answer.append(0)

    return answer
```
