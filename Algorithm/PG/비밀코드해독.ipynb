```python
def solution(n, q, ans):
    answer_list = []
    for i in range(1,n+1):
        for j in range(i+1,n+1):
            for k in range(j+1, n+1):
                for l in range(k+1, n+1):
                    for m in range(l+1, n+1):
                        code = list([i,j,k,l,m])
                        flag = 0
                        for x in range(len(q)):
                            count = 0
                            for y in range(len(q[x])):
                                if q[x][y] in code:
                                    count += 1

                            if ans[x] != count:
                                break

                            flag += 1

                        if flag == len(ans):
                            answer_list.append(code)

    return len(answer_list)
```

```python
n = 15
q = [[2, 3, 9, 12, 13], [1, 4, 6, 7, 9], [1, 2, 8, 10, 12], [6, 7, 11, 13, 15], [1, 4, 10, 11, 14]]
ans = [2,1,3,0,1]
```

```python
for g, cnt in zip(q, ans):
    print(g,cnt)
```

```python
# 다른사람 풀이
import itertools

def solution(n, q, ans):
    f = list(itertools.combinations(range(1, n + 1), 5))

    for g, cnt in zip(q, ans):
         f = [code for code in f if len(set(code) & set(g)) == cnt]

    return len(f)
```
