```python
def solution(data, ext, val_ext, sort_by):
    if ext == 'code':
        ext = 0
    elif ext == 'date':
        ext = 1
    elif ext == 'maximum':
        ext = 2
    else:
        ext = 3

    if sort_by == 'code':
        sort_by = 0
    elif sort_by == 'date':
        sort_by = 1
    elif sort_by == 'maximum':
        sort_by = 2
    else:
        sort_by = 3

    data_ext = [x for x in data if x[ext] < val_ext]

    answer = sorted(data_ext, key = lambda x:x[sort_by])

    return answer
```
