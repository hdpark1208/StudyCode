```python
from itertools import permutations
import re

def solution(expression):
    answer = 0
    operator_list = ['+','-','*']

    digit_tokens = re.findall(r'\d+', expression)
    oper_tokens = re.findall(r'[+\-*]', expression)

    for prior_list in permutations(operator_list):
        temp_digit = digit_tokens
        temp_oper = oper_tokens

        for target in prior_list:
            temp_digit, temp_oper = calc(temp_digit, temp_oper, target)

        # print(temp_digit)
        if abs(temp_digit[0]) > answer:
            answer = abs(temp_digit[0])

    return answer

def calc(digits, opers, target):
    new_digits = [int(digits[0])]
    new_opers = []

    for i in range(len(opers)):
        if opers[i] == target:
            if target == '+':
                new_digits[-1] += int(digits[i+1])
            elif target == '-':
                new_digits[-1] -= int(digits[i+1])
            elif target == '*':
                new_digits[-1] *= int(digits[i+1])
        else:
            new_digits.append(int(digits[i+1]))
            new_opers.append(opers[i])

    return new_digits, new_opers
```
