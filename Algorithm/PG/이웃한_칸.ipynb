## 코드

```python
def solution(board, h, w):
    d = [(1,0), (-1,0), (0,1), (0,-1)]
    max_h = len(board)
    max_w = len(board[0])
    color = board[h][w]
    answer = 0

    for dx, dy in d:
        if h + dx >= 0 and h + dx <= (max_h - 1) and w + dy >= 0 and w + dy <= (max_w - 1):
            if board[h + dx][w + dy] == color:
                answer += 1

    return answer
```

```python
b = [["blue", "red", "orange", "red"], ["red", "red", "blue", "orange"], ["blue", "orange", "red", "red"], ["orange", "orange", "red", "blue"]]
h = 1
w = 1
```

```python
solution(b,h,w)
```
