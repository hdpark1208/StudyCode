## 구상

- 덩치로 순위를 매긴다. 키, 몸무게로 평가하는데, 모두 높을 경우는 순위가 결정되지만 둘 중 하나씩 높다면 같은 순위로 평가한다.
- 같은 순위가 3명보다 많을 순 없을 것이다 *

## 코드

```python
next_box(6, 15)
```

```python
def next_box(w, num):
    k = 1
    while True:
        if num > (k-1) * w and num <= k*w:
            break
        k += 1

    return num + 2*(k*w - num)+1

def solution(n, w, num):
    answer = 0
    while True:
        if num <= n:
            answer += 1
            num = next_box(w, num)
        else:
            break
    return answer
```

```python
solution(13,3,6)
```
