## 코드

```python
def solution(info, n, m):
    dp = {0:0}

    for a_trace, b_trace in info:
        new_dp = dict()

        for a_sum, b_sum in dp.items():
            new_a = a_sum + a_trace
            if new_a < n:
                if new_a not in new_dp or new_dp[new_a] > b_sum:
                    new_dp[new_a] = b_sum

            new_b = b_sum + b_trace
            if new_b < m:
                if a_sum not in new_dp or new_dp[a_sum] > new_b:
                    new_dp[a_sum] = new_b

        dp = new_dp

    if not dp:
        return -1

    return min(dp.keys())
```
