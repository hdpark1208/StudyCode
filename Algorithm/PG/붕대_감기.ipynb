## 코드

```python
from collections import deque

def solution(bandage, health, attacks):
    hill_time = bandage[0]
    hill_sec = bandage[1]
    hill_bonus = bandage[2]
    max_health = health

    t = 0
    t_continue = 0
    t_limit = attacks[-1][0]

    attacks = deque(attacks)

    while health > 0 and t < t_limit:
        t += 1

        if attacks and attacks[0][0] == t:
            a_time, a_damage = attacks.popleft()
            t_continue = 0
            health -= a_damage
            # print('{}초 공격'.format(a_time))

        else:
            t_continue += 1
            if t_continue == hill_time:
                health += hill_sec + hill_bonus
                t_continue = 0
            else:
                health += hill_sec

            if health > max_health:
                health = max_health

        # print('{}초 현재 체력:{} 연속 성공: {}'.format(t, health, t_continue))

    if health <= 0:
        return -1
    else:
        return health
```
