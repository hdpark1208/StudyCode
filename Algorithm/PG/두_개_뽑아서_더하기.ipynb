## 코드

```python
def solution(numbers):
    answer = []
    numbers.sort()

    for i in range(len(numbers)):
        for j in range(i+1, len(numbers)):

            answer.append(numbers[i]+numbers[j])

    answer = set(answer)
    answer = list(answer)

    # answer = sorted(set(answer))

    return answer
```

```python
# numbers = [2,1,3,4,1]
numbers = [0,0,100,100]
```

```python
solution(numbers)
```
