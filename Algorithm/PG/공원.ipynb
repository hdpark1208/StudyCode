```python
def replace_zero(park):
    return [["1" if x != "-1" else "-1" for x in row ] for row in park]

def find_max_mats(x, y, park): # GPT 수정
    if park[x][y] == '1':
        return 0

    L_h = len(park)
    L_w = len(park[0])
    size = 1  # 최소 크기 1부터
    while x + size <= L_h and y + size <= L_w:  # 정사각형이 범위 안에 들어오는 동안
        for nx in range(size):
            for ny in range(size):
                if park[x + nx][y + ny] == '1':
                    return size - 1
        size += 1

    return size - 1


def solution(mats, park):
    L_h = len(park)
    L_w = len(park[0])
    max_size = 0
    park = replace_zero(park)
    for x in range(L_h):
        for y in range(L_w):
            cur_size = find_max_mats(x,y,park)
            if cur_size > max_size:
                max_size = cur_size

    mats.sort()
    answer = -1
    for s in mats:
        if s <= max_size:
            answer = s

    return answer
```
