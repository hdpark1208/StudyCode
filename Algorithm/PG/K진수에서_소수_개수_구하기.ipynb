## 구상

## 코드

```python
import math

def convert_base(n,k):
    digits = []
    while n > 0:
        remainder = n % k

        if remainder < 10:
            digits.append(str(remainder))
        else:
            digits.append(chr(ord('A') + remainder - 10))

        n //= k

    return ''.join(reversed(digits))

def decision_prime(n):
    if n == 1:
        return False

    if '0' in str(n):
        return False

    for i in range(2,math.floor(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True


def solution(n, k):
    answer = 0
    convert_num = convert_base(n,k)
    num_list = list(map(int, [x for x in convert_num.split("0") if x != ""]))
    for n in num_list:
        if decision_prime(n):
            answer += 1

    return answer
```
