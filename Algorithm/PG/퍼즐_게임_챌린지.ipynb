```python
# 시간초과 후 이분 탐색 적용

def solution(diffs, times, limit):
    left = 1
    right = max(diffs)
    answer = right
    while left <= right:
        mid = (left + right) // 2
        level = mid
        prev_time = 0
        total_time = 0

        for i in range(len(diffs)):

            cur_time = times[i]
            if level < diffs[i]:
                total_time += (diffs[i] - level) * (prev_time + cur_time) + cur_time

            else:
                total_time += cur_time


            prev_time = cur_time

        if total_time <= limit:
            answer = level
            right = mid - 1
        else:

            left = mid + 1

    return answer
```
