## 구상

- 중복을 피하기 위해 report를 list -> set으로
- 카운트를 위해 dict 형태 만들어서 돌리고
- k 이상인 ID 리스트를 만들고
- 해당 ID 리스트의 원소를 신고한 ID에 해당하는 처리 결과 메일 횟수 추가

## 코드

```python
def solution(id_list, report, k):
    report = set(report)
    id_dict = {id : [] for id in id_list}
    for names in report:
        reporter, reported = names.split(" ")
        id_dict[reported].append(reporter)

    result_mail = [reporter for reported, reporter in id_dict.items() if len(reporter) >= k]
    reporter_list = []

    answer_dict = {id : 0 for id in id_list}
    for rl in result_mail:
        for name in rl:
            answer_dict[name] += 1

    answer = [v for k, v in answer_dict.items()]
    return answer
```

```python
id_list = ["muzi", "frodo", "apeach", "neo"]
report = ["muzi frodo","apeach frodo","frodo neo","muzi neo","apeach muzi"]
k = 2
```

```python
solution(id_list, report, k)
```
