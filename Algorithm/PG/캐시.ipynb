## 코드

```python
from collections import deque

def solution(cacheSize, cities):
    cities = [c.upper() for c in cities]
    dq = deque([])
    answer = 0
    cur_size = 0

    if cacheSize == 0:
        return 5 * len(cities)

    for city in cities:
        if city in dq:
            answer += 1
            dq.remove(city)
            dq.append(city)

        else:
            answer += 5

            if cur_size < cacheSize:
                dq.append(city)
                cur_size += 1

            else:
                dq.popleft()
                dq.append(city)


    return answer
```
