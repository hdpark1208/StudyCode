## 구상

- key : 차량 번호, value : 출입 시간을 가지는 차량 출입 딕셔너리 2개를 만들어서

## 코드

```python
import math
from collections import defaultdict

def time_to_min(time):
    h, m = map(int, time.split(":"))
    return 60 * h + m

def solution(fees, records):
    in_dic = defaultdict(list)
    out_dic = defaultdict(list)
    fee_dic = {}

    std_time = fees[0]
    std_fee = fees[1]
    unit_time = fees[2]
    unit_fee = fees[3]

    for record in records:
        time, car_num, state = record.split(" ")
        if state == 'IN':
            in_dic[car_num].append(time_to_min(time))
        else:
            out_dic[car_num].append(time_to_min(time))

    for num, in_time_list in in_dic.items():
        cost = 0
        time = 0

        if num in out_dic:
            out_time_list = out_dic[num]
        else:
            out_time_list = [1439] # 자정

        in_time_list.sort(reverse=True)
        out_time_list.sort(reverse=True)

        while in_time_list:

            in_time = in_time_list.pop()
            if out_time_list:
                out_time = out_time_list.pop()
            else:
                out_time = 1439

            time += out_time - in_time

        if time <= std_time:
                cost += std_fee

        else:
            cost += std_fee + math.ceil((time - std_time) / unit_time) * unit_fee

        fee_dic[num] = cost

    answer = [fee_dic[key] for key in sorted(fee_dic.keys())]

    return answer
```
