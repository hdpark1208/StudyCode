```python
from collections import defaultdict

def solution(friends, gifts):
    # 1. 기록 구조 초기화
    gift_log = defaultdict(lambda: defaultdict(int))
    given = defaultdict(int)
    received = defaultdict(int)

    # 2. 기록 누적
    for record in gifts:
        giver, receiver = record.split()
        gift_log[giver][receiver] += 1
        given[giver] += 1
        received[receiver] += 1

    # 3. 선물 지수 계산
    gift_score = {}
    for friend in friends:
        gift_score[friend] = given[friend] - received[friend]

    # 4. 다음 달 받을 선물 개수 계산
    next_gift = defaultdict(int)
    for i in range(len(friends)):
        for j in range(i + 1, len(friends)):
            A, B = friends[i], friends[j]
            A_to_B = gift_log[A][B]
            B_to_A = gift_log[B][A]

            if A_to_B > B_to_A:
                next_gift[A] += 1
            elif A_to_B < B_to_A:
                next_gift[B] += 1
            else:
                if gift_score[A] > gift_score[B]:
                    next_gift[A] += 1
                elif gift_score[B] > gift_score[A]:
                    next_gift[B] += 1
                # else: 둘 다 받지 않음 (동점)

    # 5. 최대값 반환
    return max(next_gift.values(), default=0)
```
