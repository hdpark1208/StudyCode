```python
def add_score(type_, choice):
    choice -= 4
    if choice < 0:
        return type_[0], abs(choice)
    elif choice > 0:
        return type_[1], abs(choice)
    else:
        return 0, 0

def solution(survey, choices):
    answer = ''
    table = {"R":0, "T":0, "C":0, "F":0, "J":0, "M":0, "A":0, "N":0}

    for i in range(len(survey)):
        type_, score = add_score(survey[i], choices[i])
        if type_ != 0:
            table[type_] += score

    # if table['R'] >= table['T']:
    #     answer += 'R'
    # else:
    #     answer += 'T'

    # if table['C'] >= table['F']:
    #     answer += 'C'
    # else:
    #     answer += 'F'

    # if table['J'] >= table['M']:
    #     answer += 'J'
    # else:
    #     answer += 'M'

    # if table['A'] >= table['N']:
    #     answer += 'A'
    # else:
    #     answer += 'N'

    for a, b in [('R', 'T'), ('C', 'F'), ('J', 'M'), ('A', 'N')]:
        answer += a if table[a] >= table[b] else b


    return answer
```
