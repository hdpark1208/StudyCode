```python
# GPT 수정 답안

def today_to_num(today):
    yyyy, mm, dd = map(int, today.split("."))
    return yyyy, mm, dd

def date_to_days(y, m, d):
    return y * 12 * 28 + m * 28 + d

def solution(today, terms, privacies):
    term_dict = {k: int(v) for k, v in (term.split() for term in terms)}
    today_yyyy, today_mm, today_dd = today_to_num(today)
    today_days = date_to_days(today_yyyy, today_mm, today_dd)

    answer = []
    for idx, entry in enumerate(privacies):
        day, tp = entry.split()
        yyyy, mm, dd = today_to_num(day)

        # 해당 약관의 유효기간을 달 단위로 더하고 하루 빼기
        mm += term_dict[tp]
        dd -= 1
        if dd == 0:
            dd = 28
            mm -= 1
        if mm <= 0:
            yyyy -= (abs(mm) // 12 + 1)
            mm = 12 - (abs(mm) % 12)

        expiry_days = date_to_days(yyyy, mm, dd)

        if expiry_days < today_days:
            answer.append(idx + 1)

    return answer
```

```python
# 약관이 12달 이상일 수 있는점을 간과하고 더럽게 코드를 짰다

def today_to_num(today):
    yyyy, mm, dd = map(int, today.split("."))
    return yyyy, mm, dd

def destruction(terms, privacies):
    result = []
    terms_dict = {key: int(value) for key, value in (item.split() for item in terms)}

    for priv in privacies:
        day, tp = priv.split(" ")
        yyyy, mm, dd = today_to_num(day)

        mm += terms_dict[tp]
        dd -= 1

        if dd == 0:
            mm -= 1
            dd = 28

        if mm > 12:
            mm -= 12
            yyyy += 1

        if mm < 10:
            mm = "0" + str(mm)
        if dd < 10:
            dd = "0" + str(dd)


        result.append(int(str(yyyy)+str(mm)+str(dd)))

    return result

def solution(today, terms, privacies):
    answer = []
    yyyy, mm, dd = today_to_num(today)
    if mm > 12:
        mm -= 12
        yyyy += 1

    if mm < 10:
        mm = "0" + str(mm)
    if dd < 10:
        dd = "0" + str(dd)
    limit = int(str(yyyy)+str(mm)+str(dd))

    privacies_list = destruction(terms, privacies)
    for i in range(len(privacies_list)):

        if privacies_list[i] < limit:
            answer.append(i+1)

    return answer
```
