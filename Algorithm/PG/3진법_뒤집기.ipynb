```python
def solution(n):
    reversed_ternary = to_ternary(n)
    del_0 = del_zero(reversed_ternary)
    answer = ternary_to_decimal(del_0)

    return answer

def to_ternary(n):
    if n == 0:
        return "0"

    result = ""
    while n > 0:
        remainder = n % 3
        result += str(remainder)
        n = n // 3   # 몫으로 갱신

    # return result[::-1]
    return result

def del_zero(n):
    while True:
        if n[0] == '0':
            n = n[1:]
        else:
            break

    return n

def ternary_to_decimal(n):
    k = len(n)
    result = 0
    for i in range(k-1,-1,-1):
        result += int(n[k-i-1]) * (3 ** i)

    return result
```
