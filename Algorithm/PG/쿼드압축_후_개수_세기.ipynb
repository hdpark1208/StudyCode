## 구상

- 각 구간 내에서 돌면서 압축 가능 여부와 된다면 그 숫자 파악하기
- 압축이 안된 구간은 계속 진행, 된 구간은 멈춰야 하는데... 이 부분을 어떻게

## 코드

```python
def solution(arr):
    count = [0, 0]
    n = len(arr)

    def dfs(x, y, size):
        first = arr[x][y]
        for i in range(x, x + size):
            for j in range(y, y + size):
                if arr[i][j] != first:
                    half = size // 2
                    dfs(x, y, half)
                    dfs(x, y + half, half)
                    dfs(x + half, y, half)
                    dfs(x + half, y + half, half)
                    return

        # 여기까지 왔다는 건 압축 가능
        count[first] += 1

    dfs(0, 0, n)
    return count
```
