```python
from itertools import product

def solution(users, emoticons):
    discount_rates = [10, 20, 30, 40]
    x = len(emoticons)  # 이모티콘 개수
    all_discount_cases = list(product(discount_rates, repeat=x))

    answer_list = []
    for dc_tuple in all_discount_cases: # etc. (10, 20, 20, 30)
        total_price = 0
        plus_members = 0
        price_list = list(p * (100 - d) // 100 for d, p in zip(dc_tuple, emoticons))

        for dc, limit_price in users:
            purchase = 0
            purchase_index_list = [i for i in range(x) if dc_tuple[i] >= dc]

            for index in purchase_index_list:
                purchase += price_list[index]


            if purchase >= limit_price:
                plus_members += 1
            else:
                total_price += purchase


        answer_list.append([plus_members, total_price])

    answer_list = sorted(answer_list, key= lambda x: (x[0],x[1]), reverse= True)


    return answer_list[0]
```
