## 구상

- 올바른 괄호에 대한 판단을...
- 스택을 활용하여, 여는 괄호가 나온다면 추가, 닫는 괄호가 나온다면 최근 스택과 페어 되는지 체크 후 된다면 스택에 pop 시키기.
- 위 과정을 회전시키면서 반복

## 코드

```python
def solution(s):
    answer = 0
    for _ in range(len(s)):
        if is_valid(s):
            answer += 1
        s = s[1:] + s[0]  # 회전
    return answer

def is_valid(s):
    stack = []
    pairs = {')':'(', '}':'{', ']':'['}
    for c in s:
        if c in pairs.values():  # '(', '{', '['
            stack.append(c)
        else:
            if not stack or stack[-1] != pairs[c]:
                return False
            stack.pop()
    return not stack
```
