## 코드

```python
def solution(absolutes, signs):
    answer = 0

    for i in range(len(absolutes)):
        if signs[i] == True:
            answer += absolutes[i]
        else:
            answer -= absolutes[i]


    return answer
```

```python
abs , sig = [4,7,12], [True,False,True]
```

```python
solution(abs, sig)
```
