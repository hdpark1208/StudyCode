```python
# 시간, 메모리 초과 1

def solution(n, left, right):
    answer = []
    arr = [[max(i,j) for j in range(1, n+1)] for i in range(1, n+1)]
    for row in arr:
        for i in row:
            answer.append(i)


    return answer[left:right + 1]
```

```python
# 시간, 메모리 초과 2
def solution(n, left, right):
    answer = []
    arr = [[max(i,j) for j in range(1, n+1)] for i in range(1, n+1)]
    for k in range(left, right + 1):
        answer.append(arr[k // n][k % n])

    return answer
```

```python
# 정답
def solution(n, left, right):
    answer = []
    for k in range(left, right + 1):
        i, j = k // n, k % n
        answer.append(max(i + 1, j + 1))
    return answer
```

### 메모

* 수학적으로 접근!
* arr[i][j] = max(i+1, j+1) 이고 i = k // n , j = k % n 이 된다
