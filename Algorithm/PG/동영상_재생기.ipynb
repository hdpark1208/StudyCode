## 코드

```python
def m_to_s(time):
    mm, ss = map(int, time.split(":"))
    return mm * 60 + ss

def s_to_m(sec):
    mm = sec // 60
    ss = sec % 60
    if mm < 10:
        mm = "0" + str(mm)

    if ss < 10:
        ss = "0" + str(ss)

    return str(mm) + ":" + str(ss)

def solution(video_len, pos, op_start, op_end, commands):
    video_len_s = m_to_s(video_len)
    pos_s = m_to_s(pos)
    op_start_s = m_to_s(op_start)
    op_end_s = m_to_s(op_end)
    if pos_s >= op_start_s and pos_s <= op_end_s:
        pos_s = op_end_s


    for command in commands:


        if command == "next":
            if pos_s + 10 > video_len_s:
                pos_s = video_len_s
            else:
                pos_s += 10

        elif command == "prev":
            if pos_s > 10:
                pos_s -= 10
            else:
                pos_s = 0
        if pos_s >= op_start_s and pos_s <= op_end_s:
            pos_s = op_end_s


    answer = s_to_m(pos_s)
    return answer
```

```python
vl = '07:22'
p = '04:05'
os = '00:15'
oe = '04:07'
c = ['next']
```

```python
solution(vl,p,os,oe,c)
```
