##  노트북 셀들

 
아래 셀(cell)의 안쪽을 클릭하고 셀을 **실행하기** 위해서  실행버튼 혹은 *Shift+Enter*키를 누릅니다(Shift키를 누른 상태에서 Enter키를 누르면 됩니다)

```python
2+3
```

`Out[1]:  4`과 비슷한 출력메세지가 나왔다면, 잘 실행되고 있는 겁니다!

주피터노트북은 **셀들(cells)**로 구성되어 있습니다. 두 가지 유형의  셀이 중요합니다:  **마크다운 셀**과 **코드 셀**.
일반적인 글들은 이 셀처럼 마크다운 셀들에 포함됩니다. 프로그래밍 내용은 위의 `2+2` 셀처럼 코드 셀들에 포함됩니다.

처음 마크다운 셀을 선택하면 수정할 수 있는 버튼이 보여집니다.
단순히 셀을 선택한 것과 수정모드에 들어간 차이점에 주의해야 합니다.

코드 셀을 편집하려면, 셀 안쪽을 '어디든지' 클릭한 후, 원하는 대로 수정 후 다시 실행하면 됩니다 (실행버튼을 클릭하던지 *Shift+Enter*을 누루면 됩니다)
* 선택된 셀을 실행 : Ctrl + Enter
* 선택된 셀을 실행 후 다음 셀로 포커스 이동 : Shift + Enter
* 실행 후 다음줄로 이동 : Alt + Enter
* 선택모드에서 화살표 방향키 : 셀 포커스를 위 아래로 움직일 수 있음


마크다운 셀을 편집하려면, 셀 안쪽 '어디든지' 더블클릭하면 됩니다. 일반적인 글을 작성하듯 편집하면 되고 완료 후 *Shift+Enter*를 누르면 됩니다.

마크다운 셀과 코드 셀 전환 단축키:
* 마크다운 셀로 전환 : Ctrl + M M
* 코드 셀로 전환 : Ctrl + M Y

```python
#  이 셀을 마크다운 셀로 바꾸고, 한두문장 작성해보세요.
```

* 바로 윗줄(Above)에 셀 생성 : Ctrl + M A
* 바로 아랫줄(Below)에 셀 생성 : Ctrl + M B

## 주피터노트북을 위한 기본적인 마크다운 문법

파이선 코드의 결과, 의문 또는 메모 등을 마크다운 셀에 작성해야 합니다. 여기서는 마크다운을 이용해 문서를 작성하는 기본적인 사항들을 안내합니다.


새로운 문단을 시작하기 위해서는 이것처럼 빈줄을 하나 넣으면 됩니다.
 
그리고 다음처럼 여러 typeface들, 목록, 제목, 링크 등을 만들 수 있습니다.

### Typefaces


*기울임(italic체)*나 **굵은체**로 작성하려면 단어 양쪽을 \*이나 \*\*를 감싸면 됩니다. 이 셀을 더블클릭해서 확인해보세요. 

### 목록 만들기

차례가 있는 목록을 만들기 위해서는 다음처럼 숫자로 시작하는 문장을 넣으면 됩니다.

1.  첫번째 줄
2.  두번째 줄
3.  세번째 줄

차례가 없는 목록을 만들려면 `-`로 시작하는 문장을 만들면 됩니다. (몇몇 다른 기호도 가능).  
- 첫번째 줄
- 두번째 줄
- 세번째 줄
* 네번째 줄


 
들여쓰기를 이용해서 목록 안에 다시 목록을 넣을 수도 있습니다. (하지만 너무 많이 사용하는 것보다는 새로운 문단을 시작하는게 낫습니다.)

- 첫번째 줄
  - 첫번째 줄안의 첫번째 줄
  - 첫번째 줄안의 두번째 줄
  1. 첫번째 줄안의 세번째 줄
    - 그 안의 다시 첫번째 줄
    
- 두번째 줄
- 세번째 줄
